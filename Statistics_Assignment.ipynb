# Q1 (What is hypothesis testing in statistics?
   -  Hypothesis testing in statistics is a method used to make decisions or draw conclusions about a population based on sample data. It involves formulating two statements: the null hypothesis (H₀), which assumes no effect or no difference, and the alternative hypothesis (H₁), which assumes there is an effect or difference. The process uses statistical tests to analyze the sample data and determine whether there is enough evidence to reject the null hypothesis. Hypothesis testing helps in reducing uncertainty while making inferences about populations. It uses concepts like test statistics, p-value, and significance level to decide the outcome. In simple words, it provides a systematic way to check if the observed results are due to chance or a real effect.

# Q2 (What is the null hypothesis, and how does it differ from the alternative hypothesis?
  - The null hypothesis (H₀) is a statistical statement that assumes there is no significant effect, difference, or relationship in the population, and any observed result is due to chance. In contrast, the alternative hypothesis (H₁ or Ha) states that there is a significant effect, difference, or relationship present. For example, if we test whether a new drug is effective, the null hypothesis would claim "the drug has no effect," while the alternative would state "the drug has an effect." The null hypothesis is tested directly, and we either reject or fail to reject it based on evidence. The alternative hypothesis is supported only if the null is rejected. In summary, the null assumes "no change," while the alternative assumes "there is change."

# Q3 (Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
   - The significance level (α) in hypothesis testing is the probability threshold used to decide whether to reject the null hypothesis. Common values are 0.05 (5%) or 0.01 (1%). It represents the risk of making a Type I error, which means rejecting the null hypothesis when it is actually true. For example, at α = 0.05, we allow a 5% chance of incorrectly rejecting the null. During testing, if the p-value is less than or equal to α, we reject the null hypothesis; otherwise, we fail to reject it. The significance level helps in controlling the error rate and sets the strictness of the test. A lower α means stronger evidence is needed against the null, while a higher α makes rejection easier. Thus, it plays a key role in deciding the final outcome of the test.

# Q4 (What are Type I and Type II errors? Give examples of each
  -   In hypothesis testing, two types of errors can occur. A Type I error happens when we reject the null hypothesis (H₀) even though it is actually true. It is also called a false positive. For example, concluding that a medicine works when in reality it does not. A Type II error occurs when we fail to reject the null hypothesis even though it is false. It is also called a false negative. For example, concluding that a medicine does not work when in reality it is effective. Type I error is controlled by the significance level (α), while Type II error is related to the power of the test (1-β). Both errors cannot be avoided completely, but researchers try to minimize them by selecting appropriate sample sizes and tests.

# Q5 (What is the difference between a Z-test and a T-test? Explain when to use each
  -  A Z-test and a T-test are both statistical methods used to test hypotheses, but they differ in their assumptions and usage. A Z-test is applied when the population variance (σ²) is known and the sample size is large (n > 30). It assumes that the data follows a normal distribution. On the other hand, a T-test is used when the population variance is unknown and the sample size is small (n ≤ 30). The T-test relies on the Student’s t-distribution, which accounts for more variability in small samples. Z-tests are more precise with large datasets, while T-tests are more flexible for real-world cases where variance is not known. In short, Z-test = large sample + known variance, T-test = small sample + unknown variance.

# Q6 (Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.
 -   A binomial distribution shows the probability of getting a fixed number of successes in a given number of trials. For example, tossing a coin 10 times with probability p=0.5 of heads.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10   # number of trials
p = 0.5  # probability of success
size = 1000  # number of experiments

# Generate binomial distribution
data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins=range(0, n+2), edgecolor='black', align='left')
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()
```

# Q7 (Question: Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.
 -   Z-test is used when population variance is known or sample size is large.

Example: Testing if the mean of a sample differs from a population mean (μ = 50).

```python
import numpy as np
from scipy import stats

# Sample data
sample = np.array([52, 49, 51, 50, 53, 48, 52, 54, 49, 50])
population_mean = 50
population_std = 2   # assume known
n = len(sample)

# Calculate sample mean
sample_mean = np.mean(sample)

# Z-statistic
z = (sample_mean - population_mean) / (population_std / np.sqrt(n))

# p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

print("Z-statistic:", z)
print("p-value:", p_value)

# Decision at 5% significance
if p_value < 0.05:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

# Q8 (Question: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
  -  A confidence interval gives a range in which the true population mean is expected to lie with 95% confidence.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Generate normal data
data = np.random.normal(loc=100, scale=15, size=200)

# Calculate mean and standard error
mean = np.mean(data)
sem = stats.sem(data)  # standard error of mean

# Confidence interval (95%)
conf_int = stats.t.interval(0.95, len(data)-1, loc=mean, scale=sem)

print("Sample Mean:", mean)
print("95% Confidence Interval:", conf_int)

# Plot histogram
plt.hist(data, bins=20, edgecolor='black')
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f"Mean={mean:.2f}")
plt.title("Normal Distribution with 95% CI")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

# Q9 (Question: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent.
  -  Z-scores show how many standard deviations each value is away from the mean.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Function to calculate Z-scores
def calculate_z_scores(data):
    z_scores = stats.zscore(data)
    return z_scores

# Example dataset
data = np.array([10, 12, 13, 15, 18, 20, 22, 25, 30])

# Calculate Z-scores
z_scores = calculate_z_scores(data)

print("Original Data:", data)
print("Z-scores:", z_scores)

# Plot histogram of Z-scores
plt.hist(z_scores, bins=5, edgecolor='black')
plt.title("Z-score Distribution")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.show()
```
