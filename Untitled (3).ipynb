#Q1. What is the fundamental idea behind ensemble techniques? How does bagging differ from boosting in terms of approach and objective?
 -Answer:
The fundamental idea behind ensemble techniques is to combine multiple individual models (called base learners) to create a more accurate and robust predictive model. The main goal is to reduce errors caused by bias, variance, or noise in a single model by leveraging the collective intelligence of several models.

Bagging (Bootstrap Aggregating) focuses on reducing variance by training several models independently on different random subsets of the data (created using bootstrapping). The final prediction is made by averaging (for regression) or voting (for classification) the outputs of all models. Examples include Random Forests.

Boosting, on the other hand, focuses on reducing bias by training models sequentially. Each new model attempts to correct the errors made by the previous ones by giving more weight to misclassified or poorly predicted samples. Popular examples include AdaBoost and Gradient Boosting.

In short, bagging builds independent learners in parallel to stabilize predictions, while boosting builds dependent learners sequentially to improve accuracy.


#Q2. Explain how the Random Forest Classifier reduces overfitting compared to a single decision tree. Mention the role of two key hyperparameters in this process.
 -Answer:
A single decision tree often overfits because it learns the training data too closely, capturing noise and small fluctuations. The Random Forest Classifier overcomes this by combining the predictions of multiple decision trees built on different random subsets of the data and features. This randomness ensures that each tree learns slightly different patterns, making the overall model more generalized and stable.

Random Forest reduces overfitting through two main ideas:

Bootstrap Sampling (Bagging): Each tree is trained on a random subset of the data, reducing dependence on any one sample.

Random Feature Selection: At each split, only a random subset of features is considered, preventing trees from becoming too similar.

Two key hyperparameters that control this process are:

n_estimators – the number of trees in the forest. More trees generally improve stability and reduce variance.

max_features – the number of features considered at each split. Limiting this adds diversity among trees, helping prevent overfitting.

# Q3. What is Stacking in ensemble learning? How does it differ from traditional bagging/boosting methods? Provide a simple example use case.
 -Answer:
Stacking (Stacked Generalization) is an ensemble learning technique that combines multiple models (called base learners) using another model, known as a meta-learner or blender, to make the final prediction. Instead of simply averaging predictions (like in bagging) or correcting previous errors (like in boosting), stacking learns how to best combine the outputs of different models for improved performance.

In stacking, several diverse models—such as Decision Trees, Logistic Regression, and SVM—are first trained on the same dataset. Their predictions are then used as inputs for the meta-learner, which learns the optimal way to combine them.

Difference from Bagging and Boosting:

Bagging: Uses multiple similar models trained independently on random subsets of data; results are averaged or voted on.

Boosting: Trains models sequentially, each focusing on correcting the errors of the previous one.

Stacking: Trains multiple diverse models and uses a new model to combine their predictions intelligently.

# Q4. What is the OOB Score in Random Forest, and why is it useful? How does it help in model evaluation without a separate validation set?
 -Answer:
The OOB (Out-of-Bag) Score is a built-in performance estimate used in Random Forests. It is calculated using the data samples that were not included in the bootstrap sample for a particular tree. Since each tree in a Random Forest is trained on a random subset (about two-thirds) of the data, the remaining one-third (the OOB samples) act as a validation set for that tree.

After all trees are trained, each sample in the dataset is predicted using only the trees where it was OOB, and the overall accuracy (or error rate) of these predictions forms the OOB Score.

Why it’s useful:

It provides an unbiased estimate of model performance without the need for a separate validation or test set.

It saves computational time and data since no additional split is required.

It helps in tuning hyperparameters effectively during training.

#Q5. Compare AdaBoost and Gradient Boosting in terms of:
#• How they handle errors from weak learners
#• Weight adjustment mechanism
#• Typical use cases
 -Answer:
Both AdaBoost and Gradient Boosting are boosting algorithms that combine multiple weak learners (usually decision trees) sequentially to create a strong model, but they differ in how they correct previous errors and update weights.

1. Handling Errors from Weak Learners

AdaBoost:
Focuses on misclassified samples from the previous model. It increases the weights of these difficult samples so that the next weak learner gives them more attention.

Gradient Boosting:
Works by minimizing the loss function directly. Each new model is trained to predict the residual errors (differences between actual and predicted values) of the previous models.

2. Weight Adjustment Mechanism

AdaBoost:
Adjusts the weights of individual training samples. Misclassified samples get higher weights, and correctly classified ones get lower weights.

Gradient Boosting:
Does not explicitly adjust sample weights. Instead, it updates the model parameters in the direction of the negative gradient of the loss function to minimize the overall error.

3. Typical Use Cases

AdaBoost:
Works well with clean, less noisy datasets and is commonly used for binary classification tasks such as spam detection or face recognition.

Gradient Boosting:
More flexible and powerful, suitable for both classification and regression problems, such as credit scoring, house price prediction, and customer churn modeling.

#Question 6:Why does CatBoost perform well on categorical features without requiring extensive preprocessing? Briefly explain its handling of categorical variables.
 -Answer:
CatBoost is specifically designed to handle categorical features efficiently without the need for manual preprocessing steps like one-hot encoding or label encoding. It performs well because it uses a unique method called “ordered target statistics” (or ordered encoding) and “ordered boosting”, which prevent overfitting and data leakage.

How CatBoost Handles Categorical Variables:

Ordered Target Statistics:
Instead of assigning fixed numeric labels, CatBoost replaces each categorical value with an average target value calculated from previous examples only, ensuring that information from future data points is not used. This maintains data integrity and prevents target leakage.

Combinations of Categorical Features:
CatBoost automatically generates new features by combining multiple categorical columns, helping the model capture complex relationships between categories.

Efficient Encoding During Training:
The algorithm converts categorical features into numerical values internally in a statistically robust way, so users don’t need to perform one-hot or label encoding manually.

#Question 7: KNN Classifier Assignment: Wine Dataset Analysis with
Optimization
Task:
1. Load the Wine dataset (sklearn.datasets.load_wine()).
2. Split data into 70% train and 30% test.
3. Train a KNN classifier (default K=5) without scaling and evaluate using:
a. Accuracy
b. Precision, Recall, F1-Score (print classification report)
4. Apply StandardScaler, retrain KNN, and compare metrics.
5. Use GridSearchCV to find the best K (test K=1 to 20) and distance metric
(Euclidean, Manhattan).

```python
# Step 1: Import Required Libraries
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Step 2: Load the Wine Dataset
wine = load_wine()
X = wine.data
y = wine.target
```

```python
# Step 3: Split Data into 70% Train and 30% Test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
# Step 4: Train KNN Classifier (Default K=5) Without Scaling
knn_default = KNeighborsClassifier(n_neighbors=5)
knn_default.fit(X_train, y_train)
y_pred_default = knn_default.predict(X_test)
```

```python
# Step 5: Evaluate Without Scaling
print("---- Without Scaling ----")
print("Accuracy:", accuracy_score(y_test, y_pred_default))
print("\nClassification Report:\n", classification_report(y_test, y_pred_default))
```

```python
# Step 6: Apply StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Step 7: Retrain KNN with Scaled Data
knn_scaled = KNeighborsClassifier(n_neighbors=5)
knn_scaled.fit(X_train_scaled, y_train)
y_pred_scaled = knn_scaled.predict(X_test_scaled)
```

```python
# Step 8: Evaluate After Scaling
print("\n---- After Scaling ----")
print("Accuracy:", accuracy_score(y_test, y_pred_scaled))
print("\nClassification Report:\n", classification_report(y_test, y_pred_scaled))
```

```python
# Step 9: Optimize K and Distance Metric using GridSearchCV
param_grid = {
    'n_neighbors': range(1, 21),
    'metric': ['euclidean', 'manhattan']
}

grid_search = GridSearchCV(KNeighborsClassifier(), param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train_scaled, y_train)
```

```python
# Step 10: Best Parameters and Evaluation
print("\n---- Grid Search Results ----")
print("Best Parameters:", grid_search.best_params_)
print("Best Cross-Validation Accuracy:", grid_search.best_score_)
```

```python
# Evaluate with Best Model on Test Data
best_knn = grid_search.best_estimator_
y_pred_best = best_knn.predict(X_test_scaled)

print("\n---- Final Evaluation (Best Model) ----")
print("Test Accuracy:", accuracy_score(y_test, y_pred_best))
print("\nClassification Report:\n", classification_report(y_test, y_pred_best))
```

```python
# Step 11: Train the Optimized KNN and Compare Results
best_knn = grid_search.best_estimator_
best_knn.fit(X_train_scaled, y_train)
y_pred_best = best_knn.predict(X_test_scaled)

acc_best = accuracy_score(y_test, y_pred_best)
print("\n---- Optimized KNN (Best Model) ----")
print("Test Accuracy:", acc_best)
print("\nClassification Report:\n", classification_report(y_test, y_pred_best))
```

```python
# Step 5: Apply StandardScaler, Retrain KNN, and Compare Metrics
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

knn_scaled = KNeighborsClassifier(n_neighbors=5)
knn_scaled.fit(X_train_scaled, y_train)
y_pred_scaled = knn_scaled.predict(X_test_scaled)

acc_scaled = accuracy_score(y_test, y_pred_scaled)
print("\n---- After Scaling ----")
print("Accuracy:", acc_scaled)
print("\nClassification Report:\n", classification_report(y_test, y_pred_scaled))
```

#Question 8 : PCA + KNN with Variance Analysis and Visualization
Task:
1. Load the Breast Cancer dataset (sklearn.datasets.load_breast_cancer()).
2. Apply PCA and plot the scree plot (explained variance ratio).
3. Retain 95% variance and transform the dataset.
4. Train KNN on the original data and PCA-transformed data, then compare
accuracy.
5. Visualize the first two principal components using a scatter plot (color by class).

```python
# Step 1: Import Required Libraries
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Step 2: Load the Breast Cancer Dataset
data = load_breast_cancer()
X = data.data
y = data.target
```

```python
# Step 3: Scale the Data (important for PCA)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Step 4: Apply PCA and Plot Scree Plot
pca = PCA()
X_pca_full = pca.fit_transform(X_scaled)

plt.figure(figsize=(8, 5))
plt.plot(np.cumsum(pca.explained_variance_ratio_) * 100, marker='o', color='b')
plt.title("Scree Plot - Cumulative Explained Variance")
plt.xlabel("Number of Principal Components")
plt.ylabel("Cumulative Explained Variance (%)")
plt.grid(True)
plt.show()
```

```python
# Step 5: Retain 95% Variance and Transform the Dataset
pca_95 = PCA(0.95)
X_pca = pca_95.fit_transform(X_scaled)
print(f"Number of Components Retaining 95% Variance: {pca_95.n_components_}")
```

```python
# Step 6: Split Data into Train/Test Sets
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.3, random_state=42, stratify=y
)
X_train_pca, X_test_pca, _, _ = train_test_split(
    X_pca, y, test_size=0.3, random_state=42, stratify=y
)
```

```python
# Step 7: Train KNN on Original Data
knn_original = KNeighborsClassifier(n_neighbors=5)
knn_original.fit(X_train, y_train)
y_pred_original = knn_original.predict(X_test)
acc_original = accuracy_score(y_test, y_pred_original)

# Step 8: Train KNN on PCA-Transformed Data
knn_pca = KNeighborsClassifier(n_neighbors=5)
knn_pca.fit(X_train_pca, y_train)
y_pred_pca = knn_pca.predict(X_test_pca)
acc_pca = accuracy_score(y_test, y_pred_pca)
```

```python
# Step 9: Compare Accuracies
print("---- Accuracy Comparison ----")
print(f"Original Data Accuracy: {acc_original:.4f}")
print(f"PCA (95% Variance) Accuracy: {acc_pca:.4f}")
```

```python
# Step 10: Visualize First Two Principal Components
plt.figure(figsize=(8, 6))
plt.scatter(
    X_pca_full[:, 0], X_pca_full[:, 1],
    c=y, cmap='coolwarm', edgecolor='k', alpha=0.7
)
plt.title("PCA Visualization - First Two Principal Components")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.colorbar(label="Class")
plt.show()
```

#Question 9:KNN Regressor with Distance Metrics and K-Value
Analysis
Task:
1. Generate a synthetic regression dataset
(sklearn.datasets.make_regression(n_samples=500, n_features=10)).
2. Train a KNN regressor with:
a. Euclidean distance (K=5)
b. Manhattan distance (K=5)
c. Compare Mean Squared Error (MSE) for both.
3. Test K=1, 5, 10, 20, 50 and plot K vs. MSE to analyze bias-variance tradeoff.

```python
# Step 1: Import Required Libraries
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Step 2: Generate Synthetic Regression Dataset
X, y = make_regression(n_samples=500, n_features=10, noise=10, random_state=42)
```

```python
# Step 3: Split Data into Train/Test Sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
# Step 4: Standardize Features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python

# Step 5a: Train KNN Regressor with Euclidean Distance (K=5)
knn_euclidean = KNeighborsRegressor(n_neighbors=5, metric='euclidean')
knn_euclidean.fit(X_train_scaled, y_train)
y_pred_euclidean = knn_euclidean.predict(X_test_scaled)
mse_euclidean = mean_squared_error(y_test, y_pred_euclidean)
```

```python
# Step 5b: Train KNN Regressor with Manhattan Distance (K=5)
knn_manhattan = KNeighborsRegressor(n_neighbors=5, metric='manhattan')
knn_manhattan.fit(X_train_scaled, y_train)
y_pred_manhattan = knn_manhattan.predict(X_test_scaled)
mse_manhattan = mean_squared_error(y_test, y_pred_manhattan)
```

```python
# Step 6: Compare MSE for Both Distance Metrics
print("---- MSE Comparison ----")
print(f"Euclidean Distance (K=5) MSE: {mse_euclidean:.4f}")
print(f"Manhattan Distance (K=5) MSE: {mse_manhattan:.4f}")
```

```python
# Step 7: Analyze Bias-Variance Tradeoff (K vs. MSE)
k_values = [1, 5, 10, 20, 50]
mse_scores = []

for k in k_values:
    knn = KNeighborsRegressor(n_neighbors=k, metric='euclidean')
    knn.fit(X_train_scaled, y_train)
    y_pred = knn.predict(X_test_scaled)
    mse_scores.append(mean_squared_error(y_test, y_pred))
```

```python
# Step 8: Plot K vs. MSE
plt.figure(figsize=(8, 5))
plt.plot(k_values, mse_scores, marker='o', color='teal', linewidth=2)
plt.title("K vs. Mean Squared Error (MSE)")
plt.xlabel("Number of Neighbors (K)")
plt.ylabel("Mean Squared Error")
plt.grid(True)
plt.show()
```

#Question 10: KNN with KD-Tree/Ball Tree, Imputation, and Real-World
Data
Task:
1. Load the Pima Indians Diabetes dataset (contains missing values).
2. Use KNN Imputation (sklearn.impute.KNNImputer) to fill missing values.
3. Train KNN using:
a. Brute-force method
b. KD-Tree
c. Ball Tree
4. Compare their training time and accuracy.
5. Plot the decision boundary for the best-performing method (use 2 most important
features).
Dataset: Pima Indians Diabetes

```python
# Step 1: Import Required Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.impute import KNNImputer
from sklearn.preprocessing import StandardScaler
import time
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 2: Load the Pima Indians Diabetes Dataset
# Dataset link: https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv
# Columns as per dataset description
columns = [
    'Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness',
    'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome'
]
data = pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv', names=columns)

print("First 5 rows of the dataset:")
print(data.head())
```

```python
# Step 3: Replace Zeros with NaN (since 0 in some columns means missing)
cols_with_missing = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
data[cols_with_missing] = data[cols_with_missing].replace(0, np.nan)
```

```python
# Step 4: Handle Missing Values using KNN Imputer
imputer = KNNImputer(n_neighbors=5)
data_imputed = pd.DataFrame(imputer.fit_transform(data), columns=columns)
```

```python
# Step 5: Split Data into Features and Target
X = data_imputed.drop('Outcome', axis=1)
y = data_imputed['Outcome']
```

```python
# Step 6: Standardize Features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Step 7: Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
```

```python
# Step 8: Train KNN using different algorithms
results = {}

for algo in ['brute', 'kd_tree', 'ball_tree']:
    start_time = time.time()
    knn = KNeighborsClassifier(n_neighbors=5, algorithm=algo)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    end_time = time.time()
    results[algo] = {'Accuracy': acc, 'Training Time': end_time - start_time}
```

```python

# Step 9: Compare Results
results_df = pd.DataFrame(results).T
print("\n---- KNN Algorithm Comparison ----")
print(results_df)
```

```python
# Step 10: Identify Best Performing Method
best_algo = results_df['Accuracy'].idxmax()
print(f"\n✅ Best Performing Algorithm: {best_algo}")
```

```python
# Step 11: Decision Boundary Plot (using two most important features)
# We'll use 'Glucose' and 'BMI' for visualization
X_vis = data_imputed[['Glucose', 'BMI']].values
y_vis = data_imputed['Outcome'].values
```

```python
# Scale selected features
X_vis_scaled = scaler.fit_transform(X_vis)
```

```python
# Train model on these 2 features using best algorithm
knn_best = KNeighborsClassifier(n_neighbors=5, algorithm=best_algo)
knn_best.fit(X_vis_scaled, y_vis)
```

```python
# Create mesh grid
x_min, x_max = X_vis_scaled[:, 0].min() - 1, X_vis_scaled[:, 0].max() + 1
y_min, y_max = X_vis_scaled[:, 1].min() - 1, X_vis_scaled[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))
```

```python
# Predict for grid points
Z = knn_best.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
```

```python
# Plot decision boundary
plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap='coolwarm')
sns.scatterplot(x=X_vis_scaled[:, 0], y=X_vis_scaled[:, 1], hue=y_vis, palette='coolwarm', s=50)
plt.title(f"KNN Decision Boundary ({best_algo.upper()})")
plt.xlabel('Glucose (Standardized)')
plt.ylabel('BMI (Standardized)')
plt.show()
```
