# Question1: What is Dimensionality Reduction? Why is it important in Machine Learning?
 -   Dimensionality Reduction is a technique used to reduce the number of input features in a dataset while keeping the important information intact. When a dataset has too many features, it becomes difficult for models to learn correctly and processing also becomes slow. High dimensions can create noise, duplicate information, and confusion for algorithms. By reducing dimensions, the data becomes cleaner, simpler, and easier to visualize. It also helps in improving model accuracy by removing irrelevant features. Many ML models work better and faster when only the most meaningful features are used. Therefore, Dimensionality Reduction is an important step to make machine learning models efficient, accurate, and easy to interpret.

# Question 2: Name and briefly describe three common dimensionality reduction techniques.
 - 1- PCA (Principal Component Analysis):
PCA converts the original features into new smaller sets of features called principal components. These components capture the maximum variance in the data so the most important information is preserved while the dataset becomes simpler.

   2-LDA (Linear Discriminant Analysis):
   LDA is mainly used for classification problems. It reduces dimensions by finding feature combinations that best separate different classes, improving both visualization and model performance.

   3-t-SNE (t-Distributed Stochastic Neighbor Embedding):
     t-SNE is used for visualization of high-dimensional data. It maps data points into 2D or 3D space while keeping similar points close and dissimilar points far apart, making patterns easy to observe.

# Question 3: What is clustering in unsupervised learning? Mention three popular clustering algorithms.
 -  Clustering is an unsupervised learning technique where data points are grouped based on their similarity without using any predefined labels. The goal is to place similar items in the same group and keep dissimilar items in different groups. It helps in understanding patterns, structures, and hidden relationships in data. Clustering is widely used in customer segmentation, image grouping, and anomaly detection. Three popular clustering algorithms are K-Means, which divides data into fixed clusters, Hierarchical Clustering, which creates a tree-like structure of clusters, and DBSCAN, which forms clusters based on density and identifies noise separately.

# Question 6: What is time series analysis? Mention two key components of time series data.
 -  Time series analysis is a method used to study data points collected over regular time intervals to understand trends, patterns, and future behavior. It helps in predicting how a variable will change over time based on previous observations. Time series data is common in fields like finance, weather forecasting, sales prediction, and sensor monitoring. Two important components of time series data are Trend, which shows the long-term increasing or decreasing movement, and Seasonality, which represents repeating patterns that occur at fixed intervals (like daily, monthly, or yearly). Together, these components help analysts understand how data behaves over time and make accurate forecasts.

# Question 7: Describe the difference between seasonality and cyclic behavior in time series.
 -  Seasonality refers to patterns in a time series that repeat at fixed and predictable intervals, such as every day, month, or year. These patterns are usually driven by factors like weather changes, festivals, or business cycles that occur regularly. On the other hand, cyclic behavior represents long-term ups and downs that do not follow a fixed or regular schedule. Cycles are often influenced by economic conditions or market trends and can last for months or years. The key difference is that seasonality is predictable and time-bound, while cycles are irregular and harder to forecast. Both patterns help in understanding and modeling the behavior of time-based data.

# Question 8: Write Python code to perform K-means clustering on a sample dataset. Include your Python code and output in the code box below.


```python
# K-Means Clustering Example
from sklearn.cluster import KMeans
import numpy as np
```

```python
# Sample dataset (6 points with 2 features)
X = np.array([
    [1, 2],
    [1, 4],
    [1, 0],
    [10, 2],
    [10, 4],
    [10, 0]
])
```

```python
# Create KMeans model (k = 2)
kmeans = KMeans(n_clusters=2, random_state=42)
```

```python
# Fit the model
kmeans.fit(X)
```

```python
# Get cluster labels
labels = kmeans.labels_
```

```python
# Get cluster centers
centers = kmeans.cluster_centers_
```

```python
print("Cluster Labels:")
print(labels)
```

```python
print("\nCluster Centers:")
print(centers)
```

# Question 9: What is inheritance in OOP? Provide a simple example in Python.
  -  Inheritance in Object-Oriented Programming (OOP) is a mechanism where one class (child class) can use the properties and methods of another class (parent class). It helps in reusing code, reducing repetition, and organizing programs in a clean structure. The child class can extend or modify the parent’s behavior without rewriting everything. Inheritance is useful for creating logical relationships between classes and making code easier to maintain.

```python
# Parent class
class Animal:
    def sound(self):
        return "Animals make different sounds"

# Child class inheriting from Animal
class Dog(Animal):
    def sound(self):
        return "Dog barks"

# Creating objects
a = Animal()
d = Dog()

print(a.sound())   # Output: Animals make different sounds
print(d.sound())   # Output: Dog barks
```

# Question 10: How can time series analysis be used for anomaly detection?
  -  Time series analysis helps detect anomalies by monitoring how data behaves over time and identifying points that break the regular pattern. Models first learn the normal trend, seasonality, and expected variations in the time-based data. When a new data point suddenly moves too high, too low, or behaves differently from the learned pattern, it is flagged as an anomaly. Techniques like moving averages, ARIMA, and LSTM predict the expected value and compare it with the actual value. Large deviations indicate unusual behavior. This method is widely used in fraud detection, server monitoring, stock analysis, and detecting faults in machines.
