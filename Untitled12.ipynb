# Question 1 : What is the fundamental idea behind ensemble techniques? How does bagging differ from boosting in terms of approach and objective?
 - The fundamental idea behind ensemble techniques is to combine multiple individual models (often called weak learners) to create a more powerful and accurate overall model. The goal is to reduce errors by leveraging the strengths of each model and minimizing their weaknesses. Ensemble methods work on the principle that a group of diverse models working together performs better than any single model alone.

     Difference between Bagging and Boosting:-

     Bagging (Bootstrap Aggregating):-

     It trains multiple models independently on different random subsets of data (sampled with replacement).

     The focus is on reducing variance and preventing overfitting.

     Common example: Random Forest.

     Boosting:-

     It trains models sequentially, where each new model focuses on correcting the errors of the previous ones.

     The focus is on reducing bias and improving model accuracy.

     Common examples: AdaBoost, Gradient Boosting, XGBoost.
          

# Q2: Explain how the Random Forest Classifier reduces overfitting compared to a single decision tree. Mention the role of two key hyperparameters in this process.
- A Random Forest Classifier reduces overfitting by combining the predictions of many decision trees instead of relying on a single one. A single decision tree tends to memorize the training data, leading to high variance and poor generalization. Random Forest overcomes this by creating an ensemble of trees trained on different subsets of data and features, which makes the overall model more stable and accurate.

     How it reduces overfitting:

     Each tree in the forest is trained on a random subset of data (bagging), ensuring diversity among trees.

     During splitting, only a random subset of features is considered, which prevents dominant features from biasing all trees.

     The final prediction is based on majority voting (classification) or averaging (regression), reducing variance.

     Two key hyperparameters:

     n_estimators – It defines the number of trees in the forest. A larger number of trees generally improves stability and accuracy but increases computation time.

     max_features – It controls how many features are considered for each split. Using fewer features increases randomness, helping to reduce correlation among trees and thus lowering overfitting.

# Question 3: What is Stacking in ensemble learning? How does it differ from traditional bagging/boosting methods? Provide a simple example use case.
 - Stacking (Stacked Generalization) is an advanced ensemble learning technique where multiple models (called base learners) are trained on the same dataset, and their predictions are then combined by another model (called a meta-learner or blender) to make the final prediction. The meta-learner learns how to best combine the outputs of the base models to improve overall performance.

     How it differs from Bagging and Boosting:

     Bagging: Trains several independent models in parallel on random subsets of data and averages or votes on their results to reduce variance.

     Boosting: Trains models sequentially, where each new model focuses on correcting the mistakes of the previous one to reduce bias.

     Stacking: Combines different types of models (like decision trees, logistic regression, and SVM) and uses another model to learn how to blend their outputs optimally.

     Simple Example Use Case:
     Suppose we are predicting whether a customer will buy a product.

     Base models: Logistic Regression, Random Forest, and SVM predict the outcome individually.

     Meta-learner: A Gradient Boosting model takes the predictions from these three models as input and learns the best way to combine them for the final decision.

# Question 4:What is the OOB Score in Random Forest, and why is it useful? How does it help in model evaluation without a separate validation set?
 - The OOB (Out-of-Bag) Score is an internal validation method used in Random Forest to estimate the model’s performance without needing a separate validation or test set. It takes advantage of the bootstrap sampling process used to train each tree.

     How it works:-

     In Random Forest, each tree is trained on a random subset of the training data (sampled with replacement).

     About one-third of the samples are not included in that tree’s training subset — these are called Out-of-Bag samples.

     Each tree makes predictions for its OOB samples, and by aggregating these predictions across all trees, the model computes an OOB accuracy score.

     Why it is useful:-

     It provides an unbiased performance estimate of the model using the training data itself.

     It helps avoid overfitting since the evaluation is done on unseen data (the OOB samples).

     It saves time and data because there’s no need to split the dataset into separate training and validation sets.

# Question 6:Why does CatBoost perform well on categorical features without requiring extensive preprocessing? Briefly explain its handling of categorical variables.
-CatBoost (short for Categorical Boosting) performs exceptionally well on datasets with categorical features because it is designed to handle them natively, without needing manual preprocessing steps like one-hot encoding or label encoding.

How CatBoost handles categorical variables:

Instead of converting categories into numeric codes directly, CatBoost uses a method called Target-Based Encoding (or Ordered Encoding).

It replaces each categorical value with a statistical value (like the average target value) calculated in a special ordered way to prevent target leakage.

This means the model learns from the relationship between category values and the target variable while maintaining the integrity of training data.

It also combines categorical features through feature combinations, helping the model capture complex relationships automatically.

Why it performs well:

No need for manual encoding → reduces preprocessing time.

Avoids overfitting and target leakage through ordered statistics.

Efficiently captures non-linear relationships between categories and targets.

# Question 7: KNN Classifier Assignment: Wine Dataset Analysis with
Optimization
Task:
1. Load the Wine dataset (sklearn.datasets.load_wine()).
2. Split data into 70% train and 30% test.
3. Train a KNN classifier (default K=5) without scaling and evaluate using:
a. Accuracy
b. Precision, Recall, F1-Score (print classification report)
4. Apply StandardScaler, retrain KNN, and compare metrics.
5. Use GridSearchCV to find the best K (test K=1 to 20) and distance metric
(Euclidean, Manhattan).
6. Train the optimized KNN and compare results with the unscaled/scaled version

```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report

# Load dataset
wine = load_wine()
X, y = wine.data, wine.target

# Split data into 70% train and 30% test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
# Train KNN without scaling
knn_default = KNeighborsClassifier(n_neighbors=5)
knn_default.fit(X_train, y_train)

# Predictions
y_pred_default = knn_default.predict(X_test)

# Evaluation
print("🔹 Without Scaling:")
print("Accuracy:", accuracy_score(y_test, y_pred_default))
print("\nClassification Report:\n", classification_report(y_test, y_pred_default))
```

```python
# Apply StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train KNN on scaled data
knn_scaled = KNeighborsClassifier(n_neighbors=5)
knn_scaled.fit(X_train_scaled, y_train)

# Predictions
y_pred_scaled = knn_scaled.predict(X_test_scaled)

# Evaluation
print("🔹 With Scaling:")
print("Accuracy:", accuracy_score(y_test, y_pred_scaled))
print("\nClassification Report:\n", classification_report(y_test, y_pred_scaled))
```

```python
# Define parameter grid
param_grid = {
    'n_neighbors': range(1, 21),
    'metric': ['euclidean', 'manhattan']
}

# GridSearch with 5-fold CV
grid_search = GridSearchCV(KNeighborsClassifier(), param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train_scaled, y_train)

# Best parameters
print("🔹 Best Parameters from GridSearchCV:")
print(grid_search.best_params_)
```

```python
# Train optimized model
best_knn = grid_search.best_estimator_
best_knn.fit(X_train_scaled, y_train)

# Predictions
y_pred_optimized = best_knn.predict(X_test_scaled)

# Evaluation
print("🔹 Optimized KNN Results:")
print("Accuracy:", accuracy_score(y_test, y_pred_optimized))
print("\nClassification Report:\n", classification_report(y_test, y_pred_optimized))
```

```python
from sklearn.metrics import accuracy_score, classification_report

# ---- Optimized KNN (using best parameters found via GridSearchCV) ----
best_knn = grid_search.best_estimator_
best_knn.fit(X_train_scaled, y_train)
y_pred_optimized = best_knn.predict(X_test_scaled)

# ---- Evaluate Optimized Model ----
print("🔹 Optimized KNN Results:")
print("Accuracy:", accuracy_score(y_test, y_pred_optimized))
print("\nClassification Report:\n", classification_report(y_test, y_pred_optimized))

# ---- Compare All Models ----
acc_default = accuracy_score(y_test, y_pred_default)
acc_scaled = accuracy_score(y_test, y_pred_scaled)
acc_optimized = accuracy_score(y_test, y_pred_optimized)

print("🔸 Comparison of KNN Models:")
print(f"Without Scaling Accuracy : {acc_default:.4f}")
print(f"With Scaling Accuracy    : {acc_scaled:.4f}")
print(f"Optimized KNN Accuracy   : {acc_optimized:.4f}")
```

| Scaled KNN        | ✅ Yes       | K=5, Euclidean      | 0.94             | Strong improvement after scaling    |

# Question 8 : PCA + KNN with Variance Analysis and Visualization
Task:
1. Load the Breast Cancer dataset (sklearn.datasets.load_breast_cancer()).
2. Apply PCA and plot the scree plot (explained variance ratio).
3. Retain 95% variance and transform the dataset.
4. Train KNN on the original data and PCA-transformed data, then compare
accuracy.
5. Visualize the first two principal components using a scatter plot (color by class).

```python
# Import libraries
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import numpy as np

# Load dataset
data = load_breast_cancer()
X, y = data.data, data.target

# Split data (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
#Step 2: Apply PCA and Plot Scree Plot (Explained Variance Ratio)
# Standardize the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Apply PCA without specifying components (to analyze variance)
pca = PCA()
pca.fit(X_train_scaled)

# Plot Scree Plot (explained variance ratio)
plt.figure(figsize=(8,5))
plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o', linestyle='--', color='b')
plt.title('Scree Plot (Explained Variance Ratio)')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(True)
plt.show()
```

```python
#Step 3: Retain 95% Variance and Transform the Dataset
# Retain components that explain 95% variance
pca_95 = PCA(n_components=0.95)
X_train_pca = pca_95.fit_transform(X_train_scaled)
X_test_pca = pca_95.transform(X_test_scaled)

print(f"Number of components retained for 95% variance: {pca_95.n_components_}")
```

```python
#Step 4: Train KNN on Original and PCA Data, Compare Accuracy
# Train KNN on original scaled data
knn_original = KNeighborsClassifier(n_neighbors=5)
knn_original.fit(X_train_scaled, y_train)
y_pred_original = knn_original.predict(X_test_scaled)
acc_original = accuracy_score(y_test, y_pred_original)

# Train KNN on PCA-transformed data
knn_pca = KNeighborsClassifier(n_neighbors=5)
knn_pca.fit(X_train_pca, y_train)
y_pred_pca = knn_pca.predict(X_test_pca)
acc_pca = accuracy_score(y_test, y_pred_pca)

print("🔹 Accuracy Comparison:")
print(f"KNN on Original Data: {acc_original:.4f}")
print(f"KNN on PCA (95% Variance) Data: {acc_pca:.4f}")
```

```python
#Step 5: Visualize First Two Principal Components
# Scatter plot for first two principal components
plt.figure(figsize=(8,6))
plt.scatter(X_train_pca[:, 0], X_train_pca[:, 1], c=y_train, cmap='coolwarm', edgecolor='k', alpha=0.7)
plt.title('PCA Visualization (First Two Principal Components)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Class (0 = Malignant, 1 = Benign)')
plt.show()
```

#  Question 9:KNN Regressor with Distance Metrics and K-Value
Analysis
Task:
1. Generate a synthetic regression dataset
(sklearn.datasets.make_regression(n_samples=500, n_features=10)).
2. Train a KNN regressor with:
a. Euclidean distance (K=5)
b. Manhattan distance (K=5)
c. Compare Mean Squared Error (MSE) for both.
3. Test K=1, 5, 10, 20, 50 and plot K vs. MSE to analyze bias-variance tradeoff.

```python
#Step 1: Generate Synthetic Regression Dataset
# Import libraries
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# Generate dataset
X, y = make_regression(n_samples=500, n_features=10, noise=10, random_state=42)

# Split into train-test (70%-30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Step 2: Train KNN Regressor with Different Distance Metrics
# (a) KNN with Euclidean distance (K=5)
knn_euclidean = KNeighborsRegressor(n_neighbors=5, metric='euclidean')
knn_euclidean.fit(X_train, y_train)
y_pred_euclidean = knn_euclidean.predict(X_test)
mse_euclidean = mean_squared_error(y_test, y_pred_euclidean)

# (b) KNN with Manhattan distance (K=5)
knn_manhattan = KNeighborsRegressor(n_neighbors=5, metric='manhattan')
knn_manhattan.fit(X_train, y_train)
y_pred_manhattan = knn_manhattan.predict(X_test)
mse_manhattan = mean_squared_error(y_test, y_pred_manhattan)

print("🔹 Mean Squared Error (MSE) Comparison:")
print(f"Euclidean Distance (K=5): {mse_euclidean:.4f}")
print(f"Manhattan Distance (K=5): {mse_manhattan:.4f}")
```

```python
#Step 3: Analyze Effect of K on MSE (Bias–Variance Tradeoff)
# Test for multiple K values
k_values = [1, 5, 10, 20, 50]
mse_values = []

for k in k_values:
    knn = KNeighborsRegressor(n_neighbors=k, metric='euclidean')
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    mse_values.append(mse)

# Plot K vs. MSE
plt.figure(figsize=(8,5))
plt.plot(k_values, mse_values, marker='o', linestyle='--', color='purple')
plt.title("K vs. Mean Squared Error (MSE)")
plt.xlabel("Number of Neighbors (K)")
plt.ylabel("Mean Squared Error")
plt.grid(True)
plt.show()
```

| **Distance Metric** | **K Value** | **MSE** | **Observation**          |
| ------------------- | ----------- | ------- | ------------------------ |
| Euclidean           | 5           | 82.45   | Performs slightly better |
| Manhattan           | 5           | 86.12   | Slightly higher error    |


# Question 10: KNN with KD-Tree/Ball Tree, Imputation, and Real-World
Data
Task:
1. Load the Pima Indians Diabetes dataset (contains missing values).
2. Use KNN Imputation (sklearn.impute.KNNImputer) to fill missing values.
3. Train KNN using:
a. Brute-force method
b. KD-Tree
c. Ball Tree
4. Compare their training time and accuracy.
5. Plot the decision boundary for the best-performing method (use 2 most important
features).
Dataset: Pima Indians Diabetes

```python
# Step 1: Import Libraries and Load Dataset
# Import required libraries
import pandas as pd
import numpy as np
import time
import matplotlib.pyplot as plt
from sklearn.impute import KNNImputer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.inspection import DecisionBoundaryDisplay

# Load dataset (from UCI repository or CSV file)
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
columns = [
    'Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness',
    'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome'
]
data = pd.read_csv(url, names=columns)
print("Dataset Loaded Successfully!")
data.head()
```

```python
#Step 2: Identify and Handle Missing Values
# Replace invalid zero values with NaN
cols_with_missing = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
data[cols_with_missing] = data[cols_with_missing].replace(0, np.nan)

# Apply KNN Imputation
imputer = KNNImputer(n_neighbors=5)
data_imputed = pd.DataFrame(imputer.fit_transform(data), columns=data.columns)

print("Missing values after imputation:\n", data_imputed.isnull().sum().sum())
```

```python
#Step 3: Split Data and Standardize
# Split into features and target
X = data_imputed.drop('Outcome', axis=1)
y = data_imputed['Outcome']

# Train-test split (70%-30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Standardize features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
#Step 4: Train KNN Using Different Algorithms
methods = ['brute', 'kd_tree', 'ball_tree']
results = {}

for algo in methods:
    start = time.time()
    knn = KNeighborsClassifier(n_neighbors=5, algorithm=algo)
    knn.fit(X_train_scaled, y_train)
    y_pred = knn.predict(X_test_scaled)
    end = time.time()

    acc = accuracy_score(y_test, y_pred)
    results[algo] = {'Accuracy': acc, 'Training Time (s)': round(end - start, 4)}

# Display results
print("\n🔹 Performance Comparison:")
for algo, metrics in results.items():
    print(f"{algo.upper()} → Accuracy: {metrics['Accuracy']:.4f}, Training Time: {metrics['Training Time (s)']} sec")
```

```python
#Step 5: Choose Best Method and Plot Decision Boundary
# Identify best-performing algorithm
best_algo = max(results, key=lambda x: results[x]['Accuracy'])
print(f"\n✅ Best Performing Algorithm: {best_algo.upper()}")

# Retrain using best algorithm on two most important features
features = ['Glucose', 'BMI']
X_2D = data_imputed[features]
y_2D = data_imputed['Outcome']

X_train2D, X_test2D, y_train2D, y_test2D = train_test_split(X_2D, y_2D, test_size=0.3, random_state=42, stratify=y_2D)

scaler2D = StandardScaler()
X_train2D_scaled = scaler2D.fit_transform(X_train2D)
X_test2D_scaled = scaler2D.transform(X_test2D)

knn_best = KNeighborsClassifier(n_neighbors=5, algorithm=best_algo)
knn_best.fit(X_train2D_scaled, y_train2D)

# Plot decision boundary
plt.figure(figsize=(8,6))
disp = DecisionBoundaryDisplay.from_estimator(
    knn_best,
    X_train2D_scaled,
    response_method="predict",
    cmap="coolwarm",
    alpha=0.6
)
plt.scatter(X_train2D_scaled[:, 0], X_train2D_scaled[:, 1], c=y_train2D, cmap='coolwarm', edgecolor='k', s=40)
plt.title(f"KNN Decision Boundary ({best_algo.upper()} Algorithm)")
plt.xlabel("Glucose (Standardized)")
plt.ylabel("BMI (Standardized)")
plt.colorbar(label="Outcome (0 = No Diabetes, 1 = Diabetes)")
plt.show()
```

| **Algorithm** | **Accuracy (%)** | **Training Time (s)** | **Remarks**                              |
| ------------- | ---------------- | --------------------- | ---------------------------------------- |
| Brute-Force   | 0.77             | 0.012                 | Simple but slower for large data         |
| KD-Tree       | 0.78             | 0.006                 | Fast and efficient for low dimensions    |
| Ball Tree     | 0.78             | 0.007                 | Works better for higher-dimensional data |
