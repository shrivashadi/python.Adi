```python
## 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
# the list.
def sum_even_numbers(numbers):
  total = 0
  for num in numbers:
    if num % 2 == 0:
      total += num
      return total
print(sum_even_numbers([1, 2, 3, 4, 5, 6]))
```

```python
## 3. Implement a Python function that takes a list of integers and returns a new list containing the squares of
#     each number.
def square_list(numbers):
    result = []
    for num in numbers:
        result.append(num ** 2)
    return result

# Example
print(square_list([1, 2, 3, 4]))
```

```python
## 4. Write a Python function that checks if a given number is prime or not from 1 to 200
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Example
num = 37
if is_prime(num):
    print("Prime number")
else:
    print("Not a prime number")
```

```python
## 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
#     terms.
class Fibonacci:
    def __init__(self, n):
        self.n = n          # number of terms
        self.a = 0
        self.b = 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.n:
            result = self.a
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return result
        else:
            raise StopIteration

# Example
fib = Fibonacci(7)

for num in fib:
    print(num)
```

```python
## 6. Write a generator function in Python that yields the powers of 2 up to a given exponent
def powers_of_two(n):
    for i in range(n + 1):
        yield 2 ** i

# Example
for val in powers_of_two(5):
    print(val)
```

```python
## 7. Implement a generator function that reads a file line by line and yields each line as a string.
import os
print(os.listdir())
def read_file(filename):
    try:
        with open(filename, "r", encoding="utf-8") as file:
            for line in file:
                yield line.strip()
    except FileNotFoundError:
        print("File not found! Check file name/path")

# Example
for line in read_file("adi.txt"):
    print(line)
```

```python
## 8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple
data = [(1, 5), (2, 3), (4, 1), (3, 2)]

sorted_data = sorted(data, key=lambda x: x[1])

print(sorted_data)
```

```python
## 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
celsius = [0, 20, 30, 40]

fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))

print(fahrenheit)
```

```python
## 10. Create a Python program that uses `filter()` to remove all the vowels from a given string
text = "aditya"

result = "".join(filter(lambda x: x.lower() not in "aeiou", text))

print(result)
```

```python
## 11) Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:
#Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
#product of the price per item and the quantity. The product should be increased by 10,- € if the value of the
#order is smaller than 100,00 €.
#Write a Python program using lambda and map

orders = [
    (1, "Book", 2, 30),
    (2, "Pen", 5, 10),
    (3, "Bag", 1, 120),
    (4, "Notebook", 3, 20)
]

result = list(map(
    lambda x: (x[0], x[2]*x[3] + 10 if x[2]*x[3] < 100 else x[2]*x[3]),
    orders
))

print(result)
```
